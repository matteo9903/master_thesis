## CPH model

```python
cox=CoxPHSurvivalAnalysis()
```

```python
param_grid = {'n_iter': [5,10,15,20,50,100]}
cv = KFold(n_splits=5, random_state=84, shuffle=True)
gcv = GridSearchCV(cox, param_grid, return_train_score=True, cv=cv)
gcv.fit(X_train_RW_immputed, Y_train_RW)

results = pd.DataFrame(gcv.cv_results_).sort_values(by='mean_test_score', ascending=False)
results.loc[:, ~results.columns.str.endswith("_time")]
```

```python
 best_param = gcv.best_params_
print(best_param)
```

```python
cox=CoxPHSurvivalAnalysis(n_iter=best_param['n_iter']
                         )
cox.fit(X_train_RW_immputed, Y_train_RW)
```

```python
cross_val=cross_val_score(cox, X_train_RW_immputed, Y_train_RW, cv=5)
print(cross_val)
print('Avergae score:',np.mean(cross_val))
```

```python
#predicting the risk score and c idnex
ci_cox_train = concordance_index_censored(Y_train_RW["DEATH_EVENT"], Y_train_RW["OS_general_month"], cox.predict(X_train_RW_immputed))
ci_cox_test = concordance_index_censored(Y_test_RW["DEATH_EVENT"], Y_test_RW["OS_general_month"], cox.predict(X_test_RW_immputed))
#External validation
ci_cox_ex_val = concordance_index_censored(Y_NP_RW["DEATH_EVENT"], Y_NP_RW["OS_general_month"], cox.predict(X_NP_RW))


print(f'The c-index CPH for train dataset is {ci_cox_train[0]:.3f}')
print(f'The c-index CPH for test dataset is {ci_cox_test[0]:.3f}')
print(f'The c-index CPH for external validation dataset is {ci_cox_ex_val[0]:.3f}')

print(f'The number of concordant pairs in train set is {ci_cox_train[1]:.3f}')
print(f'The number of concordant pairs in test set is {ci_cox_test[1]:.3f}')
print(f'The number of concordant pairs in ex val set is {ci_cox_ex_val[1]:.3f}')

print(f'The number of tied risk pairs in train set is {ci_cox_train[3]:.3f}')
print(f'The number of tied risk pairs in test set is {ci_cox_test[3]:.3f}')
print(f'The number of concordant pairs in ex val set is {ci_cox_ex_val[3]:.3f}')
```
